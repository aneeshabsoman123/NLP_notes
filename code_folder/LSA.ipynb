# Latent Semantic Analysis/SVD

```python
import nltk
```

```python
textCorpus=['Seven continent planet','Five ocean planet','Asia largest continent','Pacific Ocean Largest','Ocean saline water']
```

```python
#splitting text corpus
text_tokens=[sent.split() for sent in textCorpus]
print(text_tokens)
```

```python
#creating occurence matrix
from sklearn.feature_extraction.text import TfidfVectorizer 

transformer=TfidfVectorizer ()
tfidf=transformer.fit_transform(textCorpus)
print(tfidf)
```

It is the dimension of the occurence matrix

```python
#Applying LSA on occurence matrix throughTruncatedSVD library

from sklearn.decomposition import TruncatedSVD
svd=TruncatedSVD(n_components=3)     #to reduce to only 3 column matrix
lsa=svd.fit_transform(tfidf)

print(lsa)
```

Dimensions of occurence matrix has reduced
